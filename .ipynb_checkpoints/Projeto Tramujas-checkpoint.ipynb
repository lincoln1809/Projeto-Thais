```python
ot_rt = df[["number","opened_at","resolved_at"]].dropna()
```

```python
ot_rt
```

```python
dt_ot = pd.to_datetime(ot_rt["opened_at"], format = "%d/%m/%Y %H:%M")
dt_ot
```

```python
dt_rt = pd.to_datetime(ot_rt["resolved_at"], format = "%d/%m/%Y %H:%M")
dt_rt
```

```python
ot_rt["difference"] = dt_rt - dt_ot
ot_rt
```

```python
ot_rt.sort_values(by = "difference", ascending = False)
```

```python
ot_rt.groupby(by= "number").mean()
```

```python
dif = ot_rt[["number", "difference"]]
dif
```

```python
d = pd.to_numeric(dif.difference, downcast = "float")
d
```

```python
dif["difference"]
```

```python
d = d * 3.6e+12
d
```

```python
dif["difference in hours"] = pd.to_numeric(dif["difference"] / 3.6e+12, downcast = "float")
dif
```

```python
dif.iloc[0:2, 2].sum()
```

```python
d = d * 60
d
```

```python
dif["difference in minutes"] = pd.to_numeric(dif["difference"] / 6e+10)
dif
```

```python
dif["difference in hours"] = dif["difference in minutes"] / 60
dif
```

```python
dif = dif.drop(['difference in minutes'], axis = 1)
dif
```

```python
dif = dif.drop(['difference'], axis = 1)
dif
```

```python
group = dif.groupby(by = ['number']).mean()
group.sort_values(by = ['difference in hours'], ascending = False)
# group
```
